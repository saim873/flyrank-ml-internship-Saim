## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd
from IPython.display import Markdown, display

data_url = (
    "https://raw.githubusercontent.com/"
    "flyrank-bih/flyrank-ml-internship-starter/"
    "main/data/raw/content_refresh_anonymized.csv"
)

df = pd.read_csv(data_url)

# Keep a safe page-level slice relevant to the ranking lane.
lane_columns = [
    "content_id",
    "impressions_90d",
    "clicks_90d",
    "sessions_90d",
    "content_age_days",
    "avg_position",
    "trend_pct",
    "trend_direction",
]

available_columns = [column for column in lane_columns if column in df.columns]
lane_df = df[available_columns].copy()

# Sketch the same proxy used by the starter pipeline.
lane_df["is_declining_label"] = (
    lane_df["trend_direction"]
    .fillna("")
    .astype(str)
    .str.lower()
    .eq("down")
    .astype(int)
)

display(Markdown(f"""
### Unit of analysis

One row represents **one anonymized content page**.

The dataframe contains **{len(lane_df):,} page-level rows**. Each row has search, traffic, age, position, and trend signals that could support a review-priority ranking.

`is_declining_label` is a sketched proxy target derived from the observed `trend_direction` column.
"""))

lane_df.head(10)
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
from IPython.display import Markdown, display

display(Markdown("""
### Why ML beats a fixed rule here

A fixed rule might say:

- review every page with low clicks, or
- review every page with a negative trend.

That can be too simple because review priority may depend on several signals together. For example, a declining page with high impressions may deserve more attention than a declining page with almost no visibility. Age, sessions, clicks, average position, and trend strength may interact in ways that are difficult to express with one `if` statement.

Machine learning can combine these signals and learn a more flexible ranking pattern.

The output should remain **decision-support**. A human reviewer should decide whether to refresh, investigate, monitor, or leave the page unchanged.
"""))
```
